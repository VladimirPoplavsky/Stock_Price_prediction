```python
def get_infinity_page_html_data(chromedriver_path="./chromedriver.exe", scroll_number=1000, sleep_time=3):
    # Create selenium driver
    url = "https://finance.yahoo.com/quote/AAPL/history?period1=345427200&period2=1651795200&interval=1d&filter=history&frequency=1d&includeAdjustedClose=true"
    s = Service(chromedriver_path)
    driver = webdriver.Chrome(service=s)
    driver.get(url)

    for i in range(1,scroll_number):
        print(f"scroll_number: {i}")
        driver.execute_script("window.scrollTo(1,50000)")
        time.sleep(sleep_time)
    html_data = driver.page_source
    driver.close()

    return html_data

get_infinity_page_html_data()
```
